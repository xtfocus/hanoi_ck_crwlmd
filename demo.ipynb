```python
# parse the nodes in json and write to csv rows

import requests
import json
from bs4 import BeautifulSoup
import csv
from strip_ech_tch import *
import time
# counter
c = 0
t = 0 # number of properties

source = "source.txt"



header = ['property_name', 'lat', 'lon', 'loc1', 'loc2', 'loc3', 'loc4']

with open('output.csv', 'w', encoding='UTF8') as o:
    writer = csv.writer(o)
    writer.writerow(header)
    
    with open(source, 'r') as h:
        for url in h.readlines():
            url_stripped = strip_ech_tch(url)
            resp = requests.get(url=url_stripped)
            soup = BeautifulSoup(resp.content.decode('utf-8'), 'html5lib')
            soup = soup.get_text()
            data = json.loads(soup[5:])
    
    
            for i in range(1, len(data[0][1])):
                locs = [j[0] for j in data[0][1][i][14][183][0][0][1]]
                coords = data[0][1][i][14][9][2:]
                writer.writerow([locs[0]] + coords + locs[1:])
                t += 1
                
            c += 1
            time.sleep(.9)
        print(c, ' pages')
        print(t, 'rows')
```
